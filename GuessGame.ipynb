```python
import random

want_to_play = True

def introduction():
    print("""
  ________                                ________                       
 /  _____/ __ __   ____   ______ ______  /  _____/_____    _____   ____  
/   \  ___|  |  \_/ __ \ /  ___//  ___/ /   \  ___\__  \  /     \_/ __ \ 
\    \_\  \  |  /\  ___/ \___ \ \___ \  \    \_\  \/ __ \|  Y Y  \  ___/ 
 \______  /____/  \___  >____  >____  >  \______  (____  /__|_|  /\___  >

""")

    global right_number
    right_number = random.randint(1, 100)

    print("Welcome to the Number Guessing Game\nI'm thinking of a number between 1 and 100")

    global guess_left

    if input("Choose a difficulty. Type 'e' for easy and 'h' for hard: ") == 'h':
        guess_left = 5
    else:
        guess_left = 10

    print(f"You have {guess_left} attempts remaining to guess the number")


def compareGuess():
    global guess_left

    if chosen_number == right_number:
        print(f"You're right, i was thinking of the number {right_number}")
        guess_left = 0
        return

    guess_left -= 1

    if chosen_number > right_number:
        print("Too high")
    else:
        print("Too low")

    if guess_left == 0:
        print(f"You loose. I was thinking of the number {right_number}\n")
    else:
        print("Guess again")
        print(f"You have {guess_left} attempts remaining to guess the number")


def game():
    global chosen_number
    chosen_number = int(input("Type your guess: "))
    compareGuess()


def continueGaming():
    global want_to_play
    if input("Want to play more? Type 'y' or 'n': ") == 'n':
        want_to_play = False


while want_to_play:
    introduction()
    while guess_left != 0:
        game()
    continueGaming()
```
